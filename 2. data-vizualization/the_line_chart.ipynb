# Line Chart
A line chart, also called a line graph, connects points of data with a line. It shows how values change over time or in order. Usually, the horizontal axis (x-axis) shows time or a sequence, and the vertical axis (y-axis) shows the values we're measuring. Line charts are useful for spotting trends over time.

Let's take an example of data and try to plot a line chart on that.

```python
# Importing libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Creating a dataframe
df = pd.DataFrame(
    {
        'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
        'ProductSalesIn2023':[19, 20.5, 20, 21, 20.5, 19, 20, 21, 21.5, 21.5, 23.5, 24]
    }
)

df
```

```python
# Setting up some customizations
plt.figure(figsize=(10, 6)) # Setting the figure size
sns.set(style='whitegrid') # Setting the background theme
```

```python
# Ploting the line graph
ax1 = sns.lineplot(
                data=df,
                x='Month',
                y='ProductSalesIn2023',
                marker='o'
                )

# Adding labels and title
plt.xlabel('\nMonth')
plt.ylabel('Product Sales In 2023 (Thousand)\n')
plt.title('Product Sales Over Months\n')

# Showing the plot
plt.show()
```

**Should we label the line graph?**
\
It's always better to label the fisrt and last point for context. If line starts from 0 then don't label the fisrt point.

```python
# Ploting the line graph
ax2 = sns.lineplot(
                data=df,
                x='Month',
                y='ProductSalesIn2023',
                marker='o'
                )

# Adding labels and title
plt.xlabel('\nMonth')
plt.ylabel('Product Sales In 2023 (Thousand)\n')
plt.title('Product Sales Over Months\n')

# Indices of the points to label
indices_to_label = [0, len(df) - 1]

# Looping to label the first and last points
for i in indices_to_label:
    row = df.iloc[i]
    ax2.text(
        x=row['Month'],
        y=row['ProductSalesIn2023'] + 0.1, # Why we added 0.1 here? to set the vertical length of the label on graph
        s=f"{row['ProductSalesIn2023']}",
        horizontalalignment='center',
        size=8,
        color='black'
        )

# Showing the plot
plt.show()
```

### Multi Line Chart
When we have two or more lines to plot on line chart.

```python
# Creating an another dataframe
df1 = pd.DataFrame(
    {
        'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
        'ProductSalesIn2022':[18.2, 17.2, 19.1, 17.3, 16.6, 17.7, 19.8, 20.9, 21, 21.1, 22.2, 22.4],
        'ProductSalesIn2023':[19, 20.5, 20, 21, 20.5, 19, 20, 21, 21.5, 21.5, 23.5, 24]
    }
)

df1
```

```python
# Melting the DataFrame to long format
df_melted = pd.melt(
    df1,
    id_vars=['Month'],
    value_vars=['ProductSalesIn2022', 'ProductSalesIn2023'],
    var_name='YearlyProductSales',
    value_name='SalesInThousand'
)

# Ploting the line graph
ax2 = sns.lineplot(
                data=df_melted,
                x='Month',
                y='SalesInThousand',
                hue='YearlyProductSales',
                marker='o'
                )

# Adding labels and title
plt.xlabel('\nMonth')
plt.ylabel('Product Sales In Thousand\n')
plt.title('Product Sales Over Months\n')

# Indices of the points to label
indices_to_label = [0, len(df1) - 1]

# Looping to label the first and last points
for i in indices_to_label:
    row = df1.iloc[i]
    ax2.text(
        x=row['Month'],
        y=row['ProductSalesIn2023'] + 0.1,
        s=f"{row['ProductSalesIn2023']}",
        horizontalalignment='center',
        size=8,
        color='black'
        )

for i in indices_to_label:
    row = df1.iloc[i]
    ax2.text(
        x=row['Month'],
        y=row['ProductSalesIn2022'] + 0.1,
        s=f"{row['ProductSalesIn2022']}",
        horizontalalignment='center',
        size=8,
        color='black'
        )

# Showing the plot
plt.show()
```

**When we should avoid the line chart?**
\
If a line chart has too many lines and becomes hard to read, it's better to use a different type of chart. Options like bar charts, scatter plots, or area charts can make complex data easier to understand.

```python
# Creating an another dataframe
df2 = pd.DataFrame(
    {
        'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
        'ProductSalesIn2019':[17, 17.5, 20, 19, 16.5, 17.5, 18, 20, 21, 21.5, 22, 22.5],
        'ProductSalesIn2020':[18, 17, 19, 17, 17.5, 16.5, 19, 20, 20.5, 21, 22, 23.5],
        'ProductSalesIn2021':[17.3, 17.5, 18.3, 17.5, 16.8, 17.2, 19.4, 20.2, 22.4, 21.1, 23.2, 23.7],
        'ProductSalesIn2022':[18.2, 17.2, 19.1, 17.3, 16.6, 17.7, 19.8, 20.9, 21, 21.1, 22.2, 22.4],
        'ProductSalesIn2023':[19, 20.5, 20, 21, 20.5, 19, 20, 21, 21.5, 21.5, 23.5, 24]
    }
)
```

```python
# Melting the DataFrame to long format
df_melted = pd.melt(
    df2,
    id_vars=['Month'],
    value_vars=['ProductSalesIn2019', 'ProductSalesIn2020', 'ProductSalesIn2021', 'ProductSalesIn2022', 'ProductSalesIn2023'],
    var_name='YearlyProductSales',
    value_name='SalesInThousand'
)

# Ploting the line graph
ax3 = sns.lineplot(
                data=df_melted,
                x='Month',
                y='SalesInThousand',
                hue='YearlyProductSales',
                marker='o'
                )

# Adding labels and title
plt.xlabel('\nMonth')
plt.ylabel('Product Sales In Thousand\n')
plt.title('Product Sales Over Months\n')

# Showing the plot
plt.show()
```
