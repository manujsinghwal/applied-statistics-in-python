# Pie Chart
A pie chart is a circle divided into sectors in such a way that each sector represent a distinct group or categories. It is often used to show percentage, where the sum of the sectors equals 100%. The size of each sector is proportional to how many are in each group/category compared with the others. The larger the sector, the greater the relative popularity of that group. The number in a particular group is called the **frequency**.

**Frequency:**
It describes how many items there are in a particular group or interval. It’s like a count of how many there are.

Let's take an example and creates the pie chart on that.

```python
# Importing libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Generating the dataset
df = pd.DataFrame({
    'Genre': ['Sports', 'Strategy', 'Action', 'Shooter', 'Other'],
    'UnitsSold': [27500, 11500, 6000, 3500, 1500]
})

df.info()
```

```python
# Setting Seaborn style
sns.set(style="whitegrid")

# Creating a color palette
palette = sns.color_palette('pastel')[0:5]  # Using Seaborn color palette

# Creating the pie chart
plt.figure(figsize=(5, 5))  # Setting the figure size
plt.pie(df['UnitsSold'], labels=df['Genre'], autopct='%1.1f%%', colors=palette)

# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')

# Adding a title
plt.title('Units Sold by Genre\n')

# Show the plot
plt.show()
```

### Interpretation
*   This sector is much larger than all the others (`Sports`), which means that the frequency is highest for this category.
*   The sector here is much smaller than the others (`Other`), so this means sales are a lot lower for this category.

### So when are pie charts useful?
Pie charts can be useful if we want to compare basic proportions. It’s usually easy to tell at a glance which groups have a high frequency compared with the others. Pie charts are less useful if all the sectors have similar sizes, as it’s difficult to pick up on visible differences between the sector sizes.

Let's take an example where we have the approx similar sectors.

```python
# Generating the dataset
df_new = pd.DataFrame({
    'Genre': ['Sports', 'Strategy', 'Action', 'Shooter', 'Other'],
    'UnitsSold': [1000, 1010, 1030, 1040, 1080]
})

df_new.info()
```

```python
# Setting Seaborn style
sns.set(style="whitegrid")

# Creating a color palette
palette = sns.color_palette('pastel')[0:5]  # Using Seaborn color palette

# Creating the pie chart
plt.figure(figsize=(5, 5))  # Setting the figure size
plt.pie(df_new['UnitsSold'], labels=df_new['Genre'], colors=palette)

# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')

# Adding a title
plt.title('Units Sold by Genre\n')

# Show the plot
plt.show()
```

Now it's difficult to tell which `Genre` have highest units solds.
